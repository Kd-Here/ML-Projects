````python
# importing the image moudle from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
````

````python
# Loading image to matplotlib.image module

img = mpimg.imread('/content/Mount.jpg')
````

````python
type(img)
print(img.shape)
# For colorful image shape is shown 3 at 3rd position which represent the channels numbers,
# 1 position is row and 2nd position is the column
````

````python
#Size of image can found as:-
print(img.shape[0]*img.shape[1])
````

````python
# displaying the image from numpy array

pic = plt.imshow(img)
plt.show()
````

Resizing image with pillow lib


````python
from PIL import Image
````

````python
img = Image.open('/content/Mount.jpg')
img_resized = img.resize((200,200))
print(img_resized)
````

````python
#saving the image.
img_resized.save('MountRE.jpg')
````

````python
pic = plt.imshow(img_resized)
plt.show()


img = mpimg.imread('/content/MountRE.jpg')
print(img.shape)
````

Converting RGB to grayscale using opencv

````python
#Importing opencv lib
import cv2

img = cv2.imread('/content/MountRE.jpg')
````

````python
type(img)
img.shape
````

````python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
````

````python
print(grayscale_image.shape)
print(grayscale_image)
````

````python
pi  = plt.imshow(grayscale_image)
plt.show()
````
