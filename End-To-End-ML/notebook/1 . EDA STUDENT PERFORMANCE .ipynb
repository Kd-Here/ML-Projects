### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total score']/3
df.head()
```

```python
""" For counting the 100 values in df['reading score']  and rest this are alternative way

#1) Here value_counts() changes the DataFrame into a Series of unique values from the 'reading score' column and arranges them in descending order of their counts. From the resulting Series, we can retrieve the count of occurrences where the value was 100 by accessing the count for the specific value using [100].
count = df['reading score'].value_counts()[100]
print(count)

#2) Here comparison operation is done on each value of dataframe which is then changed in boolean indexing value where True means value == 100 False != 100, .sum() gives count of all True 
print((df['reading score'] == 100).sum())


#3) Here we used 2nd step but instead of directly using sum() we created a new dataframe of df['reading score'] == 100] which will give all the rows that have value == 100,
lastly from new dataframe we retrive column ['reading score'] & count it
df[df['reading score'] == 100]['reading score'].count()
"""


reading_full = df[df['reading score'] == 100]['average'].count()
writing_full = df[df['writing score'] == 100]['average'].count()
math_full = df[df['math score'] == 100]['average'].count()

print(f'Number of students with full marks in Reading: {reading_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Maths: {math_full}')
```

```python
"""Alternative to find count of values <= 20

(df['reading score'] <= 20).value_counts()

1) Here it creates a boolean indexing for dataframe ['reading score'] it gives True index where value is <= 20 & False where value is greater then 20 and finally we used sum to count all True values 
count = (df['reading score'] <= 20).sum()
print(count)


2) Here we create a new DataFrame by filtering the original DataFrame df using the condition df['reading score'] <= 20. The new DataFrame will contain only the rows where the value in the 'reading score' column is less than or equal to 20. Next, we apply ['reading score'].value_counts() on the filtered DataFrame to get the count of each unique value in the 'reading score' column. Finally, we use .sum() to calculate the sum of all the counts obtained from value_counts(), giving us the total count of values less than or equal to 20.
reading_less_20 = df[df['reading score'] <= 20]['reading score'].value_counts().sum()
print(reading_less_20)

3) Which is same as above instead of seeing unique values then sum we can simply count the times value <20. Here we used average you can used any column 
print(df[df['reading score'] <= 20]['average'].count())

"""


reading_less_20 = df[df['reading score'] <= 20]['average'].count()
writing_less_20 = df[df['writing score'] <= 20]['average'].count()
math_less_20 = df[df['math score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
numeric_columns = ['average', 'math score']
gender_group = df.groupby('gender')[numeric_columns].mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']

# This is just retriving values from gender_group['average'][0] like matrix :- from gender_group choose average [0] is for female 
female_scores = [gender_group['average'][0], gender_group['math score'][0]]
male_scores = [gender_group['average'][1], gender_group['math score'][1]]

X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
